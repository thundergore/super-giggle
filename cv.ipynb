```python
import matplotlib.pyplot as plt

# Data for visualization
experience = {
    "Blizzard Entertainment": {"start": 2006, "end": 2009, "role": "Specialist Game Master", "color": "blue"},
    "Jolt Online Gaming": {"start": 2009, "end": 2012, "role": "Community Manager", "color": "gold"},
    "Lionbridge": {"start": 2012, "end": 2018, "role": "Consultant", "color": "skyblue"},
    "HBO Max": {"start": 2018, "end": 2021, "role": "Technical Support", "color": "purple"},
    "HBO Max/ WarnerBros": {"start": 2021, "end": 2023, "role": "Data Analyst", "color": "orange"},
    "Schibsted": {"start": 2023, "end": 2025, "role": "Data Analyst", "color": "green"},
}

skills = {
    "Python": 84,
    "SQL": 95,
    "NLP": 79,
    "Data Visualization": 90,
    "Machine Learning": 71,
    "Data Engineering": 70,
}

# Plotting experience timeline
fig, ax = plt.subplots(figsize=(10, 6))

# Add bars for each job
for i, (company, details) in enumerate(experience.items()):
    ax.barh(i, details["end"] - details["start"], left=details["start"], label=details["role"], color=details["color"])

# Customize the timeline
ax.set_yticks(range(len(experience)))
ax.set_yticklabels(experience.keys())
ax.set_xticks(range(2006, 2026, 2))
ax.set_xlabel("Years")
ax.set_title("Experience Timeline")
ax.legend()

# This one needs some padding adjustment
plt.tight_layout(pad=2.0)  # Add padding around the plot

# Save 
#plt.savefig("Viz/xp_timeline.png")
# Show experience timeline
plt.show()

# Plotting skills as a radar chart
import numpy as np

# Skill names and values
skill_labels = list(skills.keys())
skill_values = list(skills.values())

# Radar chart setup
num_vars = len(skill_labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
skill_values += skill_values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, skill_values, color='orange', alpha=0.25)
ax.plot(angles, skill_values, color='purple', linewidth=2)
ax.set_yticks(range(0, 101, 20))
ax.set_yticklabels(["0", "20", "40", "60", "80", "100"], color="grey")
ax.set_xticks(angles[:-1])
ax.set_xticklabels(skill_labels, color="purple")
ax.set_title("Technical Skill Proficiency", y=1.1)

# This one needs some padding adjustment
plt.tight_layout(pad=2.0)  # Add padding around the plot

# Save 
#plt.savefig("Viz/xskillz_radar.png")
# Show radar chart
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Skills and tools with frequency indicating proficiency/usage
skills = {
    "Python": 84, "SQL": 95, "NLP": 79, "Tableau": 70, "Data Viz": 90, "Snowflake": 60,
    "Docker": 60, "Flyte": 50, "DBT": 50, "Hugging Face": 45, "OpenAI": 45,
    "WhisperX": 40, "Pandas": 80, "Matplotlib": 70, "Machine Learning": 75,
    "Data Engineering": 65
}

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='spring')
wordcloud.generate_from_frequencies(skills)

# Display word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Skills and Tools Word Cloud", fontsize=12)
# Save 
#plt.savefig("Viz/xkey_wordcloud.png")
# Show
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Skills used by role
data = {
    "Blizzard": {"Data Analysis": 65, "Scripting": 5, "Big Data": 80, "Data Viz": 10, "Creative Studio": 100},
    "Jolt Online": {"Data Analysis": 15, "Scripting": 10, "Big Data": 0, "Data Viz": 15, "Creative Studio": 63},
    "Consultant": {"Data Analysis": 45, "Scripting": 40, "Big Data": 15, "Data Viz": 50, "Creative Studio": 50},
    "HBO Max": {"Data Analysis": 90, "Scripting": 65, "Big Data": 90, "Data Viz": 80, "Creative Studio": 90},
    "Schibsted": {"Data Analysis": 95, "Scripting": 90, "Big Data": 85, "Data Viz": 90, "Creative Studio": 75},
}
df = pd.DataFrame(data).T

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap="Purples", fmt="d", cbar_kws={"label": "Proficiency/Usage"})
plt.title("Skills Heatmap by Role", fontsize=16)
plt.xlabel("Skills")
plt.ylabel("Roles")
# Save 
#plt.savefig("Viz/xrole_heatmap.png")
# Show
plt.show()

```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define connections between tools and skills
connections = [
    ("Python", "Pandas"), ("Python", "NLP"), ("Python", "SQL"), ("Python", "Data Engineering"),
    ("SQL", "Snowflake"), ("SQL", "DBT"), ("SQL", "Tableau"),
    ("DBT", "Snowflake"), ("DBT", "Tableau"),
    ("Pandas", "NLP"),
    ("NLP", "Hugging Face"), ("NLP", "OpenAI"),
    ("Tableau", "Data Visualization"), ("Snowflake", "Data Engineering"),
    ("Data Engineering", "Data Pipelines"), ("Data Engineering", "Flyte"),
    ("Docker", "Flyte"), ("Flyte", "NLP")
]

# Create graph
G = nx.Graph()
G.add_edges_from(connections)

# Draw network diagram
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color="orange", edge_color="black",node_size=4000, font_color="purple",font_size=12)
plt.title("Skills and Tools Network Diagram", fontsize=12)
# Save 
#plt.savefig("Viz/xtools_network.png")
# Show
plt.show()

```
